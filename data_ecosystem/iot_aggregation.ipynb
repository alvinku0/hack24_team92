```python
import pandas as pd
```

```python
df = pd.read_csv('iot_raw_data.csv')
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%m/%d/%y %H:%M')
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%m/%d/%y %H:%M')
```

```python
print(df.shape)
df.head()
```

```python
result = df.pivot_table(index=['DateTime', 'AirCode', 'AirplaneModel', 'Passengers', 'FromAirport', 'ToAirport'], 
                        columns='ItemType', 
                        values='Amount', 
                        aggfunc='sum', 
                        fill_value=0).reset_index()
```

```python
df['AirCode'] = df['AirCode'].astype('category')
df['AirplaneModel'] = df['AirplaneModel'].astype('category')
df['FromAirport'] = df['FromAirport'].astype('category')
df['ToAirport'] = df['ToAirport'].astype('category')
df['ItemType'] = df['ItemType'].astype('category')

df['Passengers'] = pd.to_numeric(df['Passengers'], errors='coerce')
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
```

```python
print(result.shape)
result.head(20)
```

```python
result.to_json('iot_aggregated.json', orient='records', lines=False)
```
